```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# input data
K = 1.2
gamtet = 1
```

```python
# harmonics
F1 = lambda S: 1-2*S+2*S**2-2/3*(K**2/(1+K**2))*S*(1-S)*(2-2*S+2*S**2)
F2 = lambda S: 2*(K**2/(1+K**2))*S*(1-S)*(1-2*S+2*(S**2)-4/3*(K**2/(1+K**2))*S*(1-S)*(3-4*S+4*S**2))
F3 = lambda S: 81/8*(K**4/(1+K**2)**2)*(S**2)*((1-S)**2)*(1-2*S+2*S**2-3/2*(K**2/(1+K**2))*S*(1-S)*(4-6*S+6*S**2))
F1,F2,F3
```

```python
N_coef = lambda m: 4*math.pi*1*m*1
```

```python
S = np.arange(0,1.0,0.01)
F1_ar = np.array([])
F2_ar = np.array([])
F3_ar = np.array([])
N1_ar = np.array([])
N2_ar = np.array([])
N3_ar = np.array([])
for i in S:
    F1_ar = np.append(F1_ar, i*F1(i)*N_coef(1))
    F2_ar = np.append(F2_ar, i*F2(i)*N_coef(2))
    F3_ar = np.append(F3_ar, i*F3(i)*N_coef(3))
    N1_ar = np.append(N1_ar, F1(i)*N_coef(1))
    N2_ar = np.append(N2_ar, F2(i)*N_coef(2))
    N3_ar = np.append(N3_ar, F3(i)*N_coef(3))
F1_ar
```

```python
# plt.plot(S, F1_ar, label="F1")
# plt.plot(S, F2_ar, label="F2")
# plt.plot(S, F3_ar, label="F3")
# plt.xscale("log")
plt.plot(S, F1_ar+F2_ar+F3_ar, label="intensity spectrum")
plt.plot(S,  N1_ar+N2_ar+N3_ar, label="photon spectrum")
plt.xlabel("$\omega/\omega_m$")
plt.ylabel("Yeild, rel.unit")
plt.legend()
```

```python
# plt.plot(S, F1_ar, label="F1")
# plt.plot(S, F2_ar, label="F2")
# plt.plot(S, F3_ar, label="F3")
# plt.xscale("log")
plt.plot(S, F1_ar*F2_ar*F3_ar)
plt.plot(S, N1_ar*N2_ar*N3_ar)
plt.xlabel("$\omega/\omega_m$")
plt.ylabel("$dN_m/dS$")
```

```python
# what if beam collimated
S_func = lambda gamtet: 1/(1+(gamtet/(1+K**2)))
gamtet_col = np.arange(0,0.5,0.01)
```

```python
S_col = np.array([])
for i in gamtet_col:
    S_col = np.append(S_col, S_func(i))
S_col
```

```python
F1_ar_col = np.array([])
F2_ar_col = np.array([])
F3_ar_col = np.array([])
N1_ar_col = np.array([])
N2_ar_col = np.array([])
N3_ar_col = np.array([])
for i in S_col:
    F1_ar_col = np.append(F1_ar_col,  i*F1(i)*N_coef(1))
    F2_ar_col = np.append(F2_ar_col,  i*F2(i)*N_coef(2))
    F3_ar_col = np.append(F3_ar_col,  i*F3(i)*N_coef(3))
    N1_ar_col = np.append(N1_ar_col, F1(i)*N_coef(1))
    N2_ar_col = np.append(N2_ar_col, F2(i)*N_coef(2))
    N3_ar_col = np.append(N3_ar_col, F3(i)*N_coef(3))
```

```python
plt.plot(S_col, F1_ar_col+F2_ar_col+F3_ar_col)
plt.plot(S_col, N1_ar_col+N2_ar_col+N3_ar_col)
```

```python
plt.plot(S, F1_ar+F2_ar+F3_ar, label="dW/dS(no collim)")
plt.plot(S_col, F1_ar_col+F2_ar_col+F3_ar_col, "*", label="$dW/dS(0<\gamma\Theta<0.5)$")
plt.plot(S, N1_ar+N2_ar+N3_ar, label="dN/dS(no collim)")
plt.plot(S_col, N1_ar_col+N2_ar_col+N3_ar_col, "x", label="$dN/dS(0<\gamma\Theta<0.5)$")
plt.xlabel("$\omega/\omega_m$")
plt.ylabel("Yeild, rel.unit")
plt.legend()
```

```python
gam = 2000
T = 4
K_val = 0.5
c = 3E8
h = 6.62607015E-34/2/math.pi 
```

```python
# spectrum with real numbers
w = lambda gamtet: (2*math.pi*gam**2*c)/(T*(1+gamtet**2+(K_val**2)/2))
```

```python
w_ar = np.array([])
gamtet_range = np.arange(0,1E5,5E3)
for i in gamtet_range:
    w_ar = np.append(w_ar, w(i))
```

```python
F1_ar_col_r = np.array([])
for i in w_ar:
    F1_ar_col_r = np.append(F1_ar_col_r, i*F1(i)*N_coef(1))
```

```python
E_range = np.arange(0, E, 0.05)
FE_range = np.array([])
NE_range = np.array([])
for i in E_range:
    FE_range = np.append(FE_range, i*F1(i)*N_coef(1))
    NE_range = np.append(NE_range, F1(i)*N_coef(1))
FE_range, E_range
```

```python
plt.plot(E_range, FE_range, label="dW/dS(no collim)")
plt.plot(E_range, NE_range, label="dN/dS(no collim)")
plt.xlabel("$E, eV$")
plt.ylabel("Yeild, rel.unit")
plt.legend()
```
