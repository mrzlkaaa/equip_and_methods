```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# constants t5
d1 = 0.2E-3
d2 = 0.4E-3
A = 1/(d1*d2)
lam = 0.5E-6
L = 1000E-3
```

```python
def transform2d(x,y):
    return np.sinc(np.sin(x/L)*d2/lam)*np.sinc(np.sin(y/L)*d1/lam)
```

```python
#plotting 3D plot
x = np.linspace(-0.01,0.01, 100)
y = np.linspace(-0.01,0.01, 100)
X, Y = np.meshgrid(x, y)
z = np.array([])
```

```python
for i,j in zip(x,y):
    z = np.append(z, transform2d(i,j))
```

```python
for i in x:
    for j in y:
        z = np.append(z, transform2d(i,j))
z.shape
```

```python
Z = transform2d(X,Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
```

```python
ax = plt.axes(projection='3d')
ax.set_zlim(0.1)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.set_zlabel('Z', fontsize=12, rotation=0)
```

```python
Z_int = Z**2
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z_int, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot(x, y, transform2d(x,y))
```

```python
x2d = transform2d(x,0)
y2d = transform2d(0,y)
```

```python
tr_ = np.linspace(-0.01,0.01, 10)
tricks = list(map(lambda X: "{:.2f}".format(X*1E3), tr_))
```

```python
# fourier transform for a single hole
fig, ax = plt.subplots(constrained_layout=True)
plt.plot(y, y2d, label="x=0")
plt.plot(x, x2d, label="y=0")
ax.set_xlabel('coordinates, mm')
ax.set_ylabel('$f(u,v)$')
plt.xticks(x_, x_tricks)
plt.legend()
```

```python
fig, ax = plt.subplots(constrained_layout=True)
plt.plot(x, y2d**2, label="x=0")
plt.plot(y, x2d**2, label="y=0")
ax.set_xlabel('coordinates, mm')
ax.set_ylabel('$|f(u,v)|^2$')
plt.xticks(x_, x_tricks)
plt.legend()
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z**2, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
```

```python
# constants t6
d12 = 0.4E-3
d22 = 0.8E-3
A = 1/(d12*d22)
lam = 0.5E-6
L2 = 2
D = 4E-3
```

```python
def transform2d_shift(x,y):
    return 2*np.sinc(np.sin(x/L2)*d22/lam)*np.sinc(np.sin(y/L2)*d12/lam)*np.cos(math.pi*x/L2*D)
```

```python
Z_shift = transform2d_shift(X,Y)
```

```python
fig = plt.figure(figsize=plt.figaspect(0.5))
ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.plot_surface(X, Y, Z_shift, rstride=1, cstride=1,
                cmap='cividis', edgecolor='none')
# ax.set_zlim(0.1, 1.6)
ax = fig.add_subplot(1, 2, 2, projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='cividis', edgecolor='none')
```

```python
x2d_shift = transform2d_shift(x,0)
y2d_shift = transform2d_shift(0,y)
```

```python
# fourier transform for a single hole
fig, ax = plt.subplots(constrained_layout=True)
plt.plot(y, y2d_shift, label="x=0")
plt.plot(x, x2d_shift, label="y=0")
ax.set_xlabel('coordinates, mm')
ax.set_ylabel('$f(u,v)$')
plt.xticks(x_, x_tricks)
plt.legend()
```
