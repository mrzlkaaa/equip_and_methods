```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# constants
d1 = 0.2E-3
d2 = 0.4E-3
A = 1/(d1*d2)
lam = 0.5E-6
L = 1000E-3
```

```python
def transform2d(x,y):
    return np.sinc(np.sin(x/L)*d1/lam)*np.sinc(np.sin(y/L)*d2/lam)
```

```python
#plotting 3D plot
x = np.linspace(-0.01,0.01, 50)
y = np.linspace(-0.01,0.01, 50)
X, Y = np.meshgrid(x, y)
z = np.array([])
```

```python
for i,j in zip(x,y):
    z = np.append(z, transform2d(i,j))
```

```python
for i in x:
    for j in y:
        z = np.append(z, transform2d(i,j))
z.shape
```

```python
Z = transform2d(X,Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
```

```python
Z_int = Z**2
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z_int, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
```
