```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks, peak_widths
from collections import defaultdict
from fractions import Fraction
import cmath
import math
```

```python
# variables
T0 = 2
T1 = 0.25
w0 = 2*math.pi/T0
A = 1/T1
N1 = 10
N2 = 25
w0
```

```python
t_unit = []
iterator = np.arange(0,5,0.05)
for i in iterator:
    if abs(i-T0) < T1/2:
        t_unit.append(i)
t_unit
```

```python
width = t_unit[-1]-t_unit[0]
```

```python
omega = np.linspace(-3*2*math.pi/width,3*2*math.pi/width, 80)
omega = np.arange(-3*2*math.pi/width,3*2*math.pi/width, 1)
# omega = np.linspace(-150,150, 300)
omega, len(omega)
```

```python
# find period of sequantial signals
times = defaultdict(list)
for n in range(1,3):
    for t in iterator:
        if abs(t-n*T0) < T1/2:
            times[n].append(t)
# times = np.array(times.values()).mean()
np.mean(np.asarray(list(times.values())), axis=0)
```

```python
t2 = t_unit[-1]
```

```python
def unit_func(w):
#     return A*width*cmath.exp(-1j*w*T0)*np.sinc(w*width/(2*math.pi))
    return A*width*np.sinc(w*width/(2*math.pi))
def unit_func_phase(w):
    return cmath.exp(-1j*w*T0)
```

```python
# using abs() refer my plot to amplitude representaion
shifted_pulse = []
shifted_pulse_phase = []
shifted_pulse_sum = []
for w in omega:
    shifted_pulse.append(unit_func(w).real)
    shifted_pulse_phase.append(unit_func_phase(w).imag)
    shifted_pulse_sum.append(abs(unit_func(w)))
    
```

```python
plt.plot(omega, shifted_pulse)
plt.xlabel("$\omega$, rad*$\mu$$s^{-1}$")
plt.ylabel("f($\omega$)")
```

```python
plt.plot(omega, shifted_pulse_phase)
plt.xlabel("$\omega$, rad*$\mu$$s^{-1}$")
plt.ylabel("$\phi$($\omega$)")
```

```python
plt.plot(omega, shifted_pulse, label="Amp")
plt.plot(omega, shifted_pulse_phase, label="Phase")
plt.plot(omega, shifted_pulse_sum, label="Sum")
plt.xlabel("$\omega$, rad*$\mu$$s^{-1}$")
plt.ylabel("$\phi$($\omega$)")
plt.legend()
```

```python
# using abs() refer my plot to amplitude representaion
shifted_pulse_amp = []
for w in omega:
    shifted_pulse_amp.append(abs(unit_func(w)))
```

```python
# plt.plot(omega, shifted_pulse_amp)
plt.bar(omega, shifted_pulse_amp, width = 1)
plt.xlabel("$\omega$, rad*$\mu$$s^{-1}$")
plt.ylabel("|f($\omega$)|")
```

```python
def cn_coef(N):
    return A*width/T0*cmath.exp(-1j*2*math.pi*N)*np.sinc(N*width/T0)
```

```python
omega2 = np.arange(-1.5*math.pi*w0, 1.5*math.pi*w0, 0.1)
omega2, len(omega2)
```

```python
def ft_N(w, N):
    return A*width*np.sinc(w*width/2)*np.sin(N*w*T0/2)/np.sin(w*T0/2)
def ft_N_sinc(w, N):
    return A*width*np.sinc(w*width/2)
def ft_N_sines(w, N):
    return np.sin(N*w*T0/2)/np.sin(w*T0/2)
```

```python
# N1
pereodic_ft_N_1 = []
pereodic_ft_N_1_sinc = []
pereodic_ft_N_1_sines = []
for w in omega2:
    pereodic_ft_N_1.append(ft_N(w,N1))
    pereodic_ft_N_1_sinc.append(ft_N(w,N1))
    pereodic_ft_N_1_sines.append(ft_N_sines(w,N1))
```

```python
# N2
pereodic_ft_N_2 = []
pereodic_ft_N_2_sinc = []
pereodic_ft_N_2_sines = []
for w in omega2:
    pereodic_ft_N_2.append(ft_N(w,N2))
    pereodic_ft_N_2_sinc.append(ft_N(w,N2))
    pereodic_ft_N_2_sines.append(ft_N_sines(w,N2))
```

```python
fig, ax = plt.subplots(constrained_layout=True)
plt.plot(omega2, pereodic_ft_N_1, label="$f_N(\omega)$")
plt.xlabel("$\omega$, rad*$\mu$$s^{-1}$")
plt.ylabel("$f(\omega)$")

plt.legend()
```

```python
fig, ax = plt.subplots(constrained_layout=True)
plt.plot(omega2, pereodic_ft_N_1_sines, label="sin(N$\omega$$T_0$/2)/sin($\omega$$T_0$/2)")
plt.xlabel("$\omega$, rad*$\mu$$s^{-1}$")
plt.ylabel("$f(\omega)$")

plt.legend()
```

```python
fig, ax = plt.subplots(constrained_layout=True)
plt.plot(omega2, pereodic_ft_N_2_sinc, label="$AT_p$sin($T_p$$\omega$/2)")
plt.plot(omega2, pereodic_ft_N_2_sines, label="sin(N$\omega$$T_0$/2)/sin($\omega$$T_0$/2)", alpha=0.5)
plt.xlabel("$\omega$, rad*$\mu$$s^{-1}$")
plt.ylabel("$f(\omega)$")

plt.legend()
```

```python
fig, ax = plt.subplots(constrained_layout=True)
plt.plot(omega2, pereodic_ft_N_2, label="$f_N(\omega)$")
plt.xlabel("$\omega$, rad*$\mu$$s^{-1}$")
plt.ylabel("$f_N(\omega)$")
plt.legend()
```

```python
fig, ax = plt.subplots(constrained_layout=True)
plt.plot(omega2, pereodic_ft_N_1_sinc, label="$AT_p$sin($T_p$$\omega$/2)")
plt.plot(omega2, pereodic_ft_N_1_sines, label="sin(N$\omega$$T_0$/2)/sin($\omega$$T_0$/2)")
plt.xlabel("$\omega$, rad*$\mu$$s^{-1}$")
plt.ylabel("$f(\omega)$")

plt.legend()
```

```python
pereodic_ft_N_12 = np.asarray(pereodic_ft_N_1)**2
pereodic_ft_N_22 = np.asarray(pereodic_ft_N_2)**2
```

```python
# find peaks and width for 1st and 2nd spectra
peaks_1N, _ = find_peaks(pereodic_ft_N_12, height=2)
peaks_2N, _ = find_peaks(pereodic_ft_N_22, height=2)
widths_1 = peak_widths(pereodic_ft_N_12, peaks_1N, rel_height=0.5)
widths_2 = peak_widths(pereodic_ft_N_22, peaks_2N, rel_height=0.5)
# FWHM_1, FWHM_2 = np.absolute(widths_1[2:, 6]).substruct(), np.absolute(wd_2[:,7]).sum()
# widths_1, wd_1
peaks_1N, widths_1
# FWHM_1
FWHM_11 = abs(np.subtract(*np.array(widths_1)[2:, 6]))
FWHM_12 = abs(np.subtract(*np.array(widths_1)[2:, 11]))
FWHM_21 = abs(np.subtract(*np.array(widths_2)[2:, 4]))
FWHM_22 = abs(np.subtract(*np.array(widths_2)[2:, 8]))
FWHM_11, FWHM_12, FWHM_21, FWHM_22
```

```python
# plot peaks and widths for 1st spectrum
plt.plot(omega2, pereodic_ft_N_12, label="$|f(\omega)|^2$")
plt.plot(omega2[peaks_1N], pereodic_ft_N_12[peaks_1N], "x", label="Peaks")
plt.xlabel("$\omega$, rad*$\mu$$s^{-1}$")
plt.ylabel("$|f(\omega)|^2$")
plt.legend()
```

```python
# plot peaks and widths for 1st spectrum
plt.plot(omega2, pereodic_ft_N_22, label="$|f(\omega)|^2$")
plt.plot(omega2[peaks_2N], pereodic_ft_N_22[peaks_2N], "x", label="Peaks")
plt.xlabel("$\omega$, rad*$\mu$$s^{-1}$")
plt.ylabel("$|f(\omega)|^2$")
plt.legend()
```

```python
# N1 = 10
pereodic_cn_coef_1 = []
cn_x_1 = np.arange(-w0*N1,w0*N1,1)
for n in cn_x_1:
    pereodic_cn_coef_1.append(abs(cn_coef(n)))
```
