```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#input data
c = 3E8
cev = 939.571550E6 # ev
dpsi = 0.1 # mrad
dphi = 0.5 # mrad
collim = 2*dpsi*dphi
Ee = 3E9 # eV
me = 9.1093837E-31 # mass of electron
E0 = (me*c**2)/1.6E-13 # rest energy of electron in Mev
R = 476 # m
I = 100E-3 # A
con_coef = 1.6E-19
beta = 99.999997099/100
gamma = 1/math.sqrt(1-beta)
```

```python
wc = 1.5*gamma**3*c/R
lamc = 4/3*math.pi*R/(gamma**3)
wc
```

```python
# function that with 99% accuracy approximate synchrotron spectrum
fy = lambda y: (1.599*(y**(1/3))-0.624*(y**(0.5))+0.125*y)*math.exp(-y)
```

```python
y = np.arange(0.01, 10, 0.01)
res_fy = np.array([])
for i in y:
    res_fy = np.append(res_fy, fy(i))
res_fy
```

```python
plt.plot(y, res_fy)
plt.ylabel("f(y)")
plt.xlabel("$\omega$/$\omega_c$")
```

```python
hp = 6.62607015E-34
hd = hp/(2*math.pi)
alpha = con_coef**2/(hd*c)
Elower = 10E3*con_coef
wlower = Elower/hd
ylower = wlower/wc
ylower, 1/ylower
```

```python
# function that with 99% accuracy approximate number of photons in spectrum
phfy = lambda y: (1.599*(y**(-2/3))-0.624*(y**(-0.5))+0.125)*math.exp(-y)
```

```python
# y_photons = np.arange(ylower, y)
# photons
phots = 8*math.pi/9*alpha*gamma*phfy(ylower)
phy = np.arange(1E-3, 100, 0.1)
res_phfy = np.array([])
for i in phy:
    res_phfy = np.append(res_phfy, phfy(i))
res_fy, phots
```

```python
# quickcheck of lamc/lam
# function that with 99% accuracy approximate number of photons in spectrum
phfy_lam = lambda y: (1.599*(y**(2/3))-0.624*(y**(0.5))+0.125)*math.exp(-y)
```

```python
phy = np.arange(1E-3, 100, 0.1)
res_phfy_lam = np.array([])
for i in phy:
    res_phfy_lam = np.append(res_phfy_lam, phfy_lam(i))
res_fy
```

```python
plt.plot(phy, res_phfy)
plt.xscale("log")
plt.yscale("log")
plt.ylabel("f(y)/y")
plt.xlabel("$\omega$/$\omega_c$")
```

```python
# angular distribution of photons
angular_dist = lambda pg: 7*(con_coef**2)*c*gamma**5*(1+(5/7*((pg**2))/(1+(pg**2))))/(16*(R**2)*(1+(pg**2))**(5/2))
```

```python
E1 = 1E3
E5 = 5E3
E10 = 10E3
E15 = 15E3
```

```python
w1 = E1*con_coef/hd
w5 = E5*con_coef/hd
w10 = E10*con_coef/hd
w15 = E15*con_coef/hd
```

```python
wwc1 = w1/wc
wwc5 = w5/wc
wwc10 = w10/wc
wwc15 = w15/wc
wwc1, wwc15
```

```python
llc1 = 1/wwc1
llc5 = 1/wwc5
llc10 = 1/wwc10
llc15 = 1/wwc15
llc1, llc15
```

```python
# photons number
pht1 = norm_coef*phfy_lam(llc1)
pht5 = norm_coef*phfy_lam(llc5)
pht10 = norm_coef*phfy_lam(llc10)
pht15 = norm_coef*phfy_lam(llc15)
```

```python
pht1, pht5, pht10, pht15
```

```python
gp = np.arange(0,4,0.1)
ang_d = np.array([])
for i in gp:
    ang_d = np.append(ang_d, angular_dist(i))
```

```python
plt.plot(gp*phfy_lam(0.2), ang_d*phfy_lam(0.2))
plt.plot(gp*phfy_lam(1), ang_d*phfy_lam(1))
```

```python
plt.plot(gp*pht1/norm_coef, ang_d*pht1, label="E = 1 keV")
plt.plot(gp*pht5/norm_coef, ang_d*pht5, label="E = 5 keV")
plt.plot(gp*pht10/norm_coef, ang_d*pht10, label="E = 10 keV")
plt.plot(gp*pht15/norm_coef, ang_d*pht15, label="E = 15 keV")
plt.ylabel("$dP/d\psi$")
plt.xlabel("$\gamma\psi$")
plt.legend()
```
